Script python criado para realizar testes estatísticos para apoio da análise


```python
import pandas as pd
doc = 'C:/Users/souza/OneDrive/Desktop/mkt/dados_marketing.csv'
df = pd.read_csv(doc, sep= ";")
```

```python
df
```

```python
colunas_campanha = [
    'Compra na Campanha 1',
    'Compra na Campanha 2',
    'Compra na Campanha 3',
    'Compra na Campanha 4',
    'Compra na Campanha 5'
]
df['Comprou'] = df[colunas_campanha].max(axis=1)
```

```python
df.duplicated().any()
```

```python
df['Total_gasto_cliente'] = df[
    ['Gasto com Alimentos', 'Gasto com Brinquedos', 'Gasto com Eletronicos',
     'Gasto com Moveis', 'Gasto com Utilidades', 'Gasto com Vestuario']
].sum(axis=1)
```

```python
df
```

Qual a relação entre escolaridade e quantidade de indivíduos?


```python
contagem = df.groupby('Escolaridade')['ID'].count()
```

```python
porcentagem= (contagem /df['ID'].count()) * 100
porcentagem.sort_values(ascending=False)
```

Pessoas com filhos ou adolescentes em casa gastam mais ou menos?

```python

gasto_total_por_grupo_filhos = df.groupby('Filhos em Casa')['Total_gasto_cliente'].sum()

gasto_total_geral = df['Total_gasto_cliente'].sum()

porcentagem_contribuicao_receita = (gasto_total_por_grupo_filhos / gasto_total_geral) * 100

print(porcentagem_contribuicao_receita.sort_values(ascending=False).apply(lambda x: f"{x:.2f}%"))
```

Correlação entre salário anual e total gasto

```python
df.dropna(inplace=True)
```

```python
from scipy.stats import pearsonr
df_limpo = df[['Salario Anual', 'Total_gasto_cliente']].dropna()

corr, p_valor = pearsonr(df_limpo['Salario Anual'], df_limpo['Total_gasto_cliente'])
print(f"Correlação: {corr:.3f}, p-valor: {p_valor:.5f}")
```

Clientes casados e com curso superior são os que mais gastam?


```python
df['grupo'] = df['Estado Civil'] + ' | ' + df['Escolaridade']
```

```python
df
```

```python
from scipy.stats import f_oneway


grupos = [dadosgrupo['Total_gasto_cliente'].values for _, dadosgrupo in df.groupby('grupo')]

f_stat, p_valor = f_oneway(*grupos)

print(f'Estatística F: {f_stat:.3f}, p-valor: {p_valor:.5f}')
```

```python
df.groupby('grupo')['Total_gasto_cliente'].mean().sort_values(ascending=False)
```

Clientes que participaram das campanhas têm salário médio mais alto?


```python
df_limpo.info()
```

```python
from scipy import stats
grupoSim = df_limpo[df_limpo['Comprou']==1]['Salario Anual']
grupoNao = df_limpo[df_limpo['Comprou']== 0]['Salario Anual']

ttestResult = stats.ttest_ind(grupoSim, grupoNao, equal_var=False)
ttestResult
```

Perfil de quem mais compra

```python
from scipy import stats
compra_escolaridade = df.groupby('Escolaridade')['Comprou'].value_counts(normalize=True).unstack()
compra_escolaridade
```

```python
from scipy.stats import chi2_contingency
tabela_contingencia_escolaridade = pd.crosstab(df['Escolaridade'], df['Comprou'])
chi2_escolaridade, p_valor_escolaridade, _, _ = chi2_contingency(tabela_contingencia_escolaridade)


print(f"Estatística Qui-Quadrado: {chi2_escolaridade:.3f}")
print(f"P-valor: {p_valor_escolaridade:.4f}")
```

```python
compra_escolaridade = df.groupby('Pais')['Comprou'].value_counts(normalize=True).unstack()
compra_escolaridade
```

```python
from scipy.stats import chi2_contingency
tabela_contingencia_pais = pd.crosstab(df['Pais'], df['Comprou'])
chi2_pais, p_valor_pais, _, _ = chi2_contingency(tabela_contingencia_pais)


print(f"Estatística Qui-Quadrado: {chi2_pais:.3f}")
print(f"P-valor: {p_valor_pais:.4f}")
```
